```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import rasterio
from rasterio.plot import show
from matplotlib.colors import Normalize
from rasterio.warp import Resampling, reproject
#from rasterio.transform import from_origin
```

```python
with rasterio.open('cropWind2.tif') as src:
    #read single band (1 channel) 
    wind_data = src.read(1)
    
    plt.figure(figsize=(8,6))
    plt.imshow(wind_data, cmap='viridis', norm= Normalize(vmin=0, vmax=10))
    plt.colorbar(label="Wind speed(m/s)")
    #plt.show()
    
    #Get transform attribute from geotiff
    transform = src.transform
    
    #convert geo coord
    lon, lat = 140.669632, 37.441064
    pixel_col, pixel_row = ~transform * (lon, lat)
    
    #Here convert to integer indices
    pixel_col, pixel_row = int(pixel_col), int(pixel_row)
    
    wind_speed_at_location = src.read(1, window=((pixel_row, pixel_row+1), (pixel_col, pixel_col+1)))
    
print(f"The wind speed at lon=140.669632, lat=37.441064 is: {wind_speed_at_location[0]}")
    #6.022609 at lon, lat = 140.732117, 37.466139
    #6.394121 at lon, lat = 140.73, 37.46
print("Number of bands: ", src.count)
    
```
